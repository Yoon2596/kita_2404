## 더 나은 결과를 얻기 위한 프롬프트 6가지 전략
대규모 언어 모델(때로는 GPT 모델이라고도 함)에서 더 나은 결과를 얻기 위한 프롬프트 전략과 전술. 여기에 설명된 방법은 더 큰 효과를 위해 때때로 조합하여 사용

명확하고 구체적인 언어 사용
- 모호한 표현 대신 구체적 단어 사용: 일반적 표현보다 구체적인 단어를 사용하여 명확히 전달합니다. 예: "빠르다" 대신 "5분 내에".
- 명확한 문장 구조 유지: 단문을 사용하여 문장이 복잡하지 않도록 합니다.

단계별 지침 제공
- 작업 분할: 작업을 여러 단계로 나누어 각 단계를 명확히 설명합니다.
- 단계별 결과 명시: 각 단계에서 기대되는 결과를 구체적으로 작성합니다.

필요한 컨텍스트 제공
- 관련 배경 정보 포함: 작업을 이해하는 데 필요한 배경 정보를 제공합니다.
- 구체적 시간/상황 명시: 언제 또는 어떤 상황에서 작업이 수행되는지 명시합니다.

예시 제공
- 긍정적/부정적 예시 포함: 올바른 예시와 잘못된 예시를 모두 제공하여 비교할 수 있게 합니다.
- 다양한 시나리오 제시: 다양한 상황에 대한 예시를 포함합니다.

부가 정보 제공
- 추가 리소스 링크 제공: 참고할 수 있는 추가 자료나 링크를 제공합니다.
- 참조 자료 목록화: 관련 자료들을 목록으로 정리하여 쉽게 접근할 수 있게 합니다.

피드백 및 조정
- 피드백 요청: 지침의 명확성을 확인하기 위해 사용자로부터 피드백을 요청합니다.
- 지침 조정: 받은 피드백을 반영하여 지침을 수정하고 개선합니다.

```python
from typing import Optional
from openai import OpenAI

client = OpenAI(api_key="")

prompts = [
    '유튜브가 탄생한 배경에 대해 200자이내로 알려줘',
    '유튜브가 빠르게 성장한 비결을 200자 이내로 알려줘',
    '유튜브가 사회에 미친 영향흘 200자 이내로 알려줘',
    '유튜브로 인한 장점 3개 단점 3개 적어줘',
    '유튜브로 인해 새로 생긴 직업을 3개 알려줘',
    '유튜버가 수익을 어떻게 창출하는지 200자이내로 알려줘',
    '유튜버로 인한 사회 문제 3개 적어줘',
    '버튜버가 뭔자 200자이내로 알려줘',
    '버튜버가 왜 인기가 있는지 200자이내로 알려줘',
    '왜 사람들이 유튜버나 버튜버가 되려는지 200자이내로 알려줘'
]

for prompt in prompts:
    completion = client.chat.completions.create(
        model = "gpt-4o-mini-2024-07-18",
        messages = [
            {"role": "system", "content": "당신은 사회학자입니다."},
            {"role": "user", "content": prompt}
        ]
    )
    print(f'프롬프트 : {prompt}')
    print(completion.choices[0].message.content)
    print("\n" + "="*100 + "\n")
```

```python

# 명확하고 구체적인 언어 사용
# - 모호한 표현 대신 구체적 단어 사용: 일반적 표현보다 구체적인 단어를 사용하여 명확히 전달합니다. 예: "빠르다" 대신 "5분 내에".
# - 명확한 문장 구조 유지: 단문을 사용하여 문장이 복잡하지 않도록 합니다.
sys_prompts_1 =  [
    "당신은 은행 업무 전문가이다."
]
user_prompts_1 = [
    "예금이나 적금을 들었을 때 좋은 점은 무엇인가요?"
]


# 단계별 지침 제공
# - 작업 분할: 작업을 여러 단계로 나누어 각 단계를 명확히 설명합니다.
# - 단계별 결과 명시: 각 단계에서 기대되는 결과를 구체적으로 작성합니다.
sys_prompts_2 =  [
    "당신은 프로그래밍 교육자이며, 초보자에게 함수 작성법을 가르치고 있다."
]
user_prompts_2 = [
    "함수를 작성하는 기본 단계를 설명해주세요."
]

# 필요한 컨텍스트 제공
# - 관련 배경 정보 포함: 작업을 이해하는 데 필요한 배경 정보를 제공합니다.
# - 구체적 시간/상황 명시: 언제 또는 어떤 상황에서 작업이 수행되는지 명시합니다.
sys_prompts_3 =  [
    "당신은 프로젝트 매니저이며, 팀에 프로젝트 일정에 대해 설명해야 한다."
]
user_prompts_3 = [
    "다음 주에 시작할 금융 상품 추천 프로젝트의 주요 마일스톤에 대해 설명해주세요."
]

# 예시 제공
# - 긍정적/부정적 예시 포함: 올바른 예시와 잘못된 예시를 모두 제공하여 비교할 수 있게 합니다.
# - 다양한 시나리오 제시: 다양한 상황에 대한 예시를 포함합니다.
sys_prompts_4 =  [
    "당신은 파이썬 수강생이다."
]
user_prompts_4 = [
    "파이썬 언어를 배울 때 코드를 따라치면서 배울 때 긍정적인 효과와 부정적인 효과를 알려주세요."
]

# 부가 정보 제공
# - 추가 리소스 링크 제공: 참고할 수 있는 추가 자료나 링크를 제공합니다.
# - 참조 자료 목록화: 관련 자료들을 목록으로 정리하여 쉽게 접근할 수 있게 합니다.
sys_prompts_5 =  [
    "당신은 은행 업무 전문가이다."
]
user_prompts_5 = [
    "예금이나 적금을 신청할 수 있는 리소스 링크를 제공해 주세요."
]

# 피드백 및 조정
# - 피드백 요청: 지침의 명확성을 확인하기 위해 사용자로부터 피드백을 요청합니다.
# - 지침 조정: 받은 피드백을 반영하여 지침을 수정하고 개선합니다.
sys_prompts_6 =  [
    "당신은 파이썬 프로젝트 팀 리더이며, 팀원들에게 코드 작업에 대한 피드백을 요청하려고 한다."
]
user_prompts_6 = [
    "파이썬 프로젝트 시안에 대해 팀원들에게 피드백을 요청하는 이메일을 작성해주세요."
]

# main
sys_prompts = []
user_prompts = []

sys_prompts.extend(sys_prompts_1 + sys_prompts_2 + sys_prompts_3 + sys_prompts_4 + sys_prompts_5 + sys_prompts_6)
user_prompts.extend(user_prompts_1 + user_prompts_2 + user_prompts_3 + user_prompts_4 + user_prompts_5 + user_prompts_6)

for sys_prompt, user_prompt in zip(sys_prompts, user_prompts):
    completion = client.chat.completions.create(
        model="gpt-4o-mini-2024-07-18",
        messages=[
            {"role": "system", "content": sys_prompt},
            {"role": "user", "content": user_prompt},
        ]
    )
    print(f"프롬프트: {user_prompt}")
    print(completion.choices[0].message.content)
    print('\n'+'='*100+'\n')
```

Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.


```python
import pandas as pd
import numpy as np

a = pd.DataFrame(np.arange(6).reshape(2, 3))
print(a)
b = pd.DataFrame(np.arange(10, 16).reshape(2, 3))
print(b)

```

```python
a.iloc[0,1] = np.nan
print(a, '\n')

# 더하기
c1 = a.add(b, fill_value=0)
print(c1,'\n')
# 뺴기
c2 = a.sub(b, fill_value=0)
print(c2, '\n')
# 곱하기
c3 = a.mul(b, fill_value=0)
print(c3,'\n')
# 나누기
c4 = round(b.div(a, fill_value=0), 2)
print(c4, '\n')
df = pd.concat([c1, c2, c3, c4], axis=1)
# df.columns = ['add', 'sub', 'mul', 'div']
print(df)
```

Task2_0528. 주어진 DataFrame의 N열에서 Series s값을 빼고 결과를 새로운 열 O에 저장 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "M": [15, 25, 35, 45, 55],
    "N": [100, 200, 300, 400, 500]
})
print(df,'\n')
# Series 생성
s = pd.Series([5, 10, 15, 20, 25])
print(s)
```

```python
# 3가지 방법
df.loc[:, 'O'] = df[df.columns[1]] - s
df.loc[:, '0'] = df['N'] - s
df.loc[:, '0'] = df[['N']].iloc[:, 0] - s
df
```

```python
df[['N']]  # 데이터프레임
df['N']   # 시리즈
df.loc[:, 'N'] # 시리즈
df.loc[:, ['N']]    # 데이터프레임
```

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})

df
```

```python
# df['A+D'] = df['A'] + df['D']
# df['Compare_E'] = df[df.E > df['A+D']],[True]
# df['Compare_E'] = df[df.E <= df['A+D'],][False]
df
# 'A+D' 열 생성
df['A+D'] = df['A'] + df['D']

# 조건에 따라 'Compare_E' 열 생성
df['Compare_E'] = np.where(df['E'] > df['A+D'], 1, 0)
df
# 또다른 방법
# 'A+D' 열 생성
df['A+D'] = df['A'] + df['D']

# 함수 정의
def compare_e(row):
    return 1 if row['E'] > row['A+D'] else 0

# 'Compare_E' 열 생성
df['Compare_E'] = df.apply(compare_e, axis=1)

df
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

# ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ë¥¼ ë¡œë”©í•©ë‹ˆë‹¤
iris = load_iris()
iris.keys()

# iris.dataëŠ” Iris ë°ì´ë„ˆ ì„¸íŠ¸ì—ì„œ í”¼ì²˜(feature)ë§Œë“œë¡œ ëœ ë°ì´í„°ë¥¼ numpyë¡œ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤
iris_data = iris.data

# iris.targetì€ ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ì—ì„œ ë ˆì´ë¸”(ê²°ì • ê°’) ë°ì´í„°ë¥¼ numpyë¡œ ê°€ì§€ê³  ìˆìŠµë‹ˆë‹¤
iris_label = iris.target
print('iris target ê°’ : ', iris_label)
print('iris target ëª… : ', iris.target_names)

# ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ë¥¼ ìì„¸íˆ ë³´ê¸° ìœ„í•´ DataFrameìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
iris.keys()
```

```python
#  DecisionTreeClassifier ê°ì²´ ìƒì„±
dt_clf = DecisionTreeClassifier(random_state=11)

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)

# í•™ìŠµ ìˆ˜í–‰
dt_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
#  DecisionTreeClassifier ê°ì²´ ìƒì„±
# dt_clf = DecisionTreeClassifier(random_state=56)

dt_clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=56)

# ë¶“ê½ƒ ë°ì´í„°ë¥¼ ë¡œë”©í•˜ê³  í•™ìŠµê³¼ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì…‹ìœ¼ë¡œ ë¶„ë¦¬
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=11)

# DecisionTreeClassifer í•™ìŠµ
dt_clf.fit(X_train, y_train)

# ì„±ëŠ¥ í‰ê°€

pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print('ê²°ì • íŠ¸ë¦¬ ì˜ˆì¸¡ ì •í™•ë„ : {0:.4f}'.format(accuracy))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# ë°ì´í„° ë¡œë“œ
cancer = load_breast_cancer()

# ë°ì´í„° ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)

# List of C variable to try
C_values = [0.01, 0.1, 1, 10, 100]
accuracy_dict = {}

# Loop through different C values
for C in C_values:
    # Created SVM Classifier
    svm = SVC(kernel='linear', C=C, random_state=42)  # kernel='linear'ì€ ì„ í˜• ì»¤ë„ì„ ì‚¬ìš©í•˜ê² ë‹¤ëŠ” ê²ƒì´ê³ , C=1.0ì€ ì •ê·œí™” ë§¤ê°œë³€ìˆ˜

    # Train the classifier
    svm.fit(X_train, y_train)

    # Make predictions
    y_pred = svm.predict(X_test)

    # Evaluate the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_dict[C] = accuracy
    print(f"C : {C}, Accuracy: {accuracy:.4f}")
```

### Voting Classifier
- í•˜ë“œ ë³´íŒ… : ë‹¤ìˆ˜ê²° ì›ì¹™, ë‹¤ìˆ˜ì˜ ë¶„ë¥˜ê¸°ê°€ ê²°ì •í•œ ì˜ˆì¸¡ê°’ì„ ìµœì¢… ë³´íŒ… ê²°ê³¼ê°’ìœ¼ë¡œ ì„ ì •

```python
import pandas as pd
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# ë°ì´í„° ë¡œë“œ
cancer = load_breast_cancer()

data_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
data_df.head()
```

```python
# ê°œë³„ ëª¨ë¸ì€ ë¡œì§€ìŠ¤í‹± íšŒê·€ì™€ KNN
lr_clf = LogisticRegression(max_iter=10000, solver='lbfgs')
knn_clf = KNeighborsClassifier(n_neighbors=7)

# ê°œë³„ ëª¨ë¸ì„ ì†Œí”„íŠ¸ ë³´íŒ… ê¸°ë°˜ì˜ ì•™ìƒë¸” ëª¨ë¸ë¡œ êµ¬í˜„í•œ ë¶„ë¥˜ê¸°
vo_clf = VotingClassifier(estimators=[('LR', lr_clf), ('KNN', knn_clf)], voting='soft')

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.2, random_state=156)

# VotingClassifier í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€
vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
# print(f'Voting ë¶„ë¥˜ê¸° ì •í™•ë„: {accuracy_score(y_test, pred):.4f}')
print('Voting ë¶„ë¥˜ê¸° ì •í™•ë„: {0: .4f}'.format(accuracy_score(y_test, pred)))

# ê°œë³„ ëª¨ë¸ì˜ í•™ìŠµ/ì˜ˆì¸¡/í‰ê°€
classifiers = [lr_clf, knn_clf]
for classifier in classifiers:
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_test)
    class_name = classifier.__class__.__name__
    # print(f'{class_name} ì •í™•ë„: {accuracy_score(y_test, pred):.4f}')
    print('{0} ì •í™•ë„ : {1: .4f}'.format(class_name, accuracy_score(y_test, pred)))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
cancer = load_breast_cancer()

# Split the dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)

# í•˜ì´í¼íŒŒë¼ë¯¸í„° ê·¸ë¦¬ë“œ ì„¤ì •
hyperparameters = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


# ëª¨ë¸ ìƒì„±
rf = RandomForestClassifier(random_state=42)

# GridSerchCV ì„¤ì •
grid_search = GridSearchCV(estimator=rf,
                           param_grid=hyperparameters,
                           cv=5,       # êµì°¨ ê²€ì¦ íšŸìˆ˜
                           n_jobs=-1,  # ëª¨ë“  CPU ì½”ì–´ ì‚¬ìš©
                           verbose=2)  # 0ì€ ë‚´ì—­ì„ ë³´ì—¬ì£¼ì§€ ì•Šê³ , 1ì€ ê°„ë‹¨í•˜ê²Œ, 2ëŠ” ìƒì„¸í•˜ê²Œ ë³´ì—¬ì¤Œ.


# ëª¨ë¸ í›ˆë ¨
grid_search.fit(X_train, y_train)

# ìµœì ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¶œë ¥
print(f"Best parameters: {grid_search.best_params_}")

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)

# ëª¨ë¸ í‰ê°€
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Confusion_matrix: {conf_matrix}")
print(f"Classification_report: {class_report}")

# ì¤‘ìš” ë³€ìˆ˜ ì‹œê°í™”
importances = best_rf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12,8))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), cancer.feature_names[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score, roc_auc_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# ëª¨ë¸ í‰ê°€
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    # ëª¨ë¸ í‰ê°€
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print()
    print(f"Confusion_matrix: {conf_matrix}")
    print(f"Classification_report: {class_report}")



# ë°ì´í„° ë¡œë“œ
cancer = load_breast_cancer()

# ë°ì´í„° ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)

# ë°ì´í„° í‘œì¤€í™”
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# í•˜ì´í¼íŒŒë¼ë¯¸í„° ê·¸ë¦¬ë“œ ì„¤ì •
hyperparameters = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


# ëª¨ë¸ ìƒì„±
rf = RandomForestClassifier(random_state=42)

# GridSerchCV ì„¤ì •
grid_search = GridSearchCV(estimator=rf,
                           param_grid=hyperparameters,
                           cv=5,       # êµì°¨ ê²€ì¦ íšŸìˆ˜
                           n_jobs=-1,  # ëª¨ë“  CPU ì½”ì–´ ì‚¬ìš©
                           verbose=2)  # 0ì€ ë‚´ì—­ì„ ë³´ì—¬ì£¼ì§€ ì•Šê³ , 1ì€ ê°„ë‹¨í•˜ê²Œ, 2ëŠ” ìƒì„¸í•˜ê²Œ ë³´ì—¬ì¤Œ.

# ëª¨ë¸ í›ˆë ¨
grid_search.fit(X_train, y_train)

# ìµœì ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¶œë ¥
print(f"Best parameters: {grid_search.best_params_}")

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_rf = grid_search.best_estimator_
# y_pred = best_rf.predict(X_test)
# y_pred_proba = best_rf.predict_proba(X_test)[:, 1]
evaluate_model(best_rf, X_train, X_test, y_train, y_test)

# ì¤‘ìš” ë³€ìˆ˜ ì‹œê°í™”
importances = best_rf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12,8))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), cancer.feature_names[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
plt.show()
```

- ë¡œì§€ìŠ¤í‹± í•¨ìˆ˜ì˜ ì •ì˜

    - f(x) = 1 / (1 + e^-x)

    - eëŠ” ìì—° ìƒìˆ˜(ì•½ 2.71828). xëŠ” ì–´ë–¤ ì‹¤ìˆ˜ ê°’ë„ ê°€ëŠ¥í•˜ë©°, -ë¬´í•œëŒ€ì—ì„œ ë¬´í•œëŒ€ê¹Œì§€ì˜ ë²”ìœ„ë¥¼ ê°€ì§€ë©° ì´ í•¨ìˆ˜ëŠ” ëª¨ë“  ì‹¤ìˆ˜ ì…ë ¥ì— ëŒ€í•´ 0ê³¼ 1 ì‚¬ì´ì˜ ê°’ì„ ë°˜í™˜.

    - í•¨ìˆ˜ê°€ ê²°ê³¼ë¥¼ 0ê³¼ 1 ì‚¬ì´ë¡œ ì œí•œí•˜ê¸° ë•Œë¬¸ì—, ì´ëŠ” í™•ë¥ ì— ëŒ€í•´ ë…¼ì˜í•  ë•Œ íŠ¹íˆ ìœ ìš©. ë¡œì§€ìŠ¤í‹± íšŒê·€ ë¶„ì„ì—ì„œ ì´ í•¨ìˆ˜ëŠ” ì„ í˜• í•¨ìˆ˜ì˜ ê²°ê³¼ë¥¼ í™•ë¥ ë¡œ ë³€í™˜í•˜ëŠ”ë° ì‚¬ìš©.

```python
import numpy as np
import matplotlib.pyplot as plt

# ë¡œì§“ í•¨ìˆ˜ ì •ì˜
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ì…ë ¥ ê°’ ë²”ìœ„
x = np.linspace(-10, 10, 1000)

# ë¡œì§€ìŠ¤í‹± í•¨ìˆ˜ ì ìš©
y = sigmoid(x)

# ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title('Logistic Regression : Logit Function')
plt.xlabel('Input value (x)')
plt.ylabel('Probability (y)')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.preprocessing import StandardScaler

# ë°ì´í„° ë¡œë“œ
cancer = load_breast_cancer()

# ë°ì´í„° ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=0)

# ë°ì´í„° í‘œì¤€í™”
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ë°ì´í„° í•™ìŠµ
lr_clf = LogisticRegression(max_iter=100, solver='lbfgs', random_state=42)
lr_clf.fit(X_train, y_train)
lr_pred = lr_clf.predict(X_test)
lr_pred_proba = lr_clf.predict_proba(X_test)[:, 1]

# ëª¨ë¸ í‰ê°€

# accuracyì™€ roc_auc ì¸¡ì •
print('accuracy : {:0.3f}'.format(accuracy_score(y_test, lr_pred)))
print('roc_auc : {:0.3f}'.format(roc_auc_score(y_test, lr_pred_proba)))
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')

param_grid = {'penalty':['l2', 'l1'],
              'C':[0.01, 0.1, 1, 5, 10]
              }

grid_clf = GridSearchCV(lr_clf, param_grid=param_grid, cv=3)
grid_clf.fit(X_train, y_train)
print('ìµœì  í•˜ì´í¼íŒŒë¼ë¯¸í„° :{0} , ìµœì  í‰ê·  ì •í™•ë„ : {1:.3f}'.format( grid_clf.best_params_, grid_clf.best_score_))
```

```python
best_lr_clf = grid_clf.best_estimator_
pred1 = best_lr_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred1)
print(f"Accuracy: {accuracy:.4f}")
```

#### ë² ì´ì¦ˆ ì •ë¦¬
ë² ì´ì¦ˆ ì •ë¦¬ëŠ” ìƒˆë¡œìš´ ì¦ê±°ê°€ ì£¼ì–´ì¡Œì„ ë•Œ ì‚¬ì „ í™•ë¥ ì„ ì–´ë–»ê²Œ ì—…ë°ì´íŠ¸í•´ì•¼ í•˜ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì¤‘ìš”í•œ í™•ë¥ ë¡ ì  ë„êµ¬ì…ë‹ˆë‹¤. ì´ëŠ” íŠ¹íˆ ì˜í•™, ê¸ˆìœµ, ë¨¸ì‹ ëŸ¬ë‹ ë“± ë‹¤ì–‘í•œ ë¶„ì•¼ì—ì„œ ë„ë¦¬ ì‚¬ìš©ë©ë‹ˆë‹¤. ë² ì´ì¦ˆ ì •ë¦¬ë¥¼ í†µí•´ ì¡°ê±´ë¶€ í™•ë¥ ì„ ê³„ì‚°í•˜ë©´ ë”ìš± ì •í™•í•œ ì˜ì‚¬ ê²°ì •ì„ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

Q. ì–´ë–¤ ë§ˆì„ ì „ì²´ ì‚¬ëŒë“¤ì˜ 10.5%ê°€ ì•” í™˜ìì´ê³ , 89.5%ê°€ ì•” í™˜ìê°€ ì•„ë‹™ë‹ˆë‹¤. ì´ ë§ˆì„ì˜ ëª¨ë“  ì‚¬ëŒì— ëŒ€í•´ ì•” ê²€ì§„ì„ ì‹¤ì‹œí–ˆë‹¤ê³  í•©ì‹œë‹¤. ì•” ê²€ì§„ì‹œ ì–‘ì„± íŒì •, ìŒì„± íŒì • ê²°ê³¼ê°€ ë‚˜ì˜¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤. í•˜ì§€ë§Œ ê²€ì§„ì´ 100% ì •í™•í•˜ì§€ëŠ” ì•Šê³  ì•½ê°„ì˜ ì˜¤ì°¨ê°€ ìˆìŠµë‹ˆë‹¤. ì•” í™˜ì ì¤‘ ì–‘ì„± íŒì •ì„ ë°›ì€ ë¹„ìœ¨ì€ 90.5%, ì•” í™˜ì ì¤‘ ìŒì„± íŒì •ì„ ë°›ì€ ë¹„ìœ¨ì€ 9.5%, ì•” í™˜ìê°€ ì•„ë‹Œ ì‚¬ëŒ ì¤‘ ì–‘ì„± íŒì •ì„ ë°›ì€ ë¹„ìœ¨ì€ 20.4%, ì•” í™˜ìê°€ ì•„ë‹Œ ì‚¬ëŒ ì¤‘ ìŒì„± íŒì •ì„ ë°›ì€ ë¹„ìœ¨ì€ 79.6%ì…ë‹ˆë‹¤. ì–´ë–¤ ì‚¬ëŒì´ ì–‘ì„± íŒì •ì„ ë°›ì•˜ì„ ë•Œ ì´ ì‚¬ëŒì´ ì•” í™˜ìì¼ í™•ë¥ ì€ ì–¼ë§ˆì¼ê¹Œìš”?

```python
# ì‚¬ì „ í™•ë¥  (P(Cancer), P(No Cancer))
P_Cancer = 0.105
P_NoCancer = 0.895

# ê²€ì‚¬ì˜ ë¯¼ê°ë„ì™€ íŠ¹ì´ë„ (P(Positive|Cancer), P(Negative|Cancer), P(Positive|NoCancer), P(Negative|NoCancer))
P_Positive_Cancer = 0.905
P_Negative_Cancer = 0.095
P_Positive_NoCancer = 0.204
P_Negative_NoCancer = 0.796

# ë² ì´ì¦ˆ ì •ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ (P(Cancer|Positive))ë¥¼ ê³„ì‚°
# P_Positive : ì „ì²´ ì¸êµ¬ì—ì„œ ì•” í™˜ìì¼ í™•ë¥ 
P_Positive = (P_Positive_Cancer * P_Cancer) + (P_Positive_NoCancer * P_NoCancer)
print(P_Positive)
P_Cancer_Positive = (P_Positive_Cancer * P_Cancer) / P_Positive

print(f"ì–‘ì„± íŒì •ì„ ë°›ì€ ì‚¬ëŒì´ ì•” í™˜ìì¼ í™•ë¥ ì€ : {round(P_Cancer_Positive, 3)*100}%")
```

Q. ì‚¬ì´í‚·ëŸ° datasetsì— ìˆëŠ” load_breast_cancer ë°ì´í„°ì— ëŒ€í•˜ì—¬ GaussianNB ì•Œê³ ë¦¬ì¦˜ì„ ì ìš©í•˜ì—¬ ë¶„ë¥˜ ë¶„ì„ ëª¨ë¸ë§ ë° í‰ê°€ë¥¼ ìˆ˜í–‰í•˜ì„¸ìš”.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# ë°ì´í„° ë¡œë“œ
cancer = load_breast_cancer()

# ë°ì´í„° ë¶„í•  : í›ˆë ¨ ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=42)

# ê°€ìš°ì‹œì•ˆ ë‚˜ì´ë¸Œ ë² ì´ì¦ˆ ëª¨ë¸ ìƒì„± ë° í›ˆë ¨
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡
y_pred = gnb.predict(X_test)

# í›ˆë ¨ëœ ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ X_testì— ëŒ€í•œ ì˜ˆì¸¡ì„ ìˆ˜í–‰
print('Prediction :', y_pred, '\n')
print('Actual :', y_test, '\n')

# ì •í™•ë„ ê³„ì‚°
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

## GBM(Gradient Boosting Machine)

GBMì˜ ì£¼ìš” ì›ë¦¬ì™€ íŠ¹ì§•

2. ìˆœì°¨ì  í•™ìŠµ:
  - GBMì˜ í•µì‹¬ì€ í•™ìŠµê¸°ë“¤ì„ ìˆœì°¨ì ìœ¼ë¡œ í•™ìŠµì‹œí‚¤ë©° ê° í•™ìŠµê¸°ëŠ” ì´ì „ í•™ìŠµê¸°ì˜ ì”ì—¬ ì˜¤ì°¨(residual error)ì— ì§‘ì¤‘í•˜ì—¬ í•™ìŠµ. ì´ëŠ” ê° í•™ìŠµê¸°ê°€ ì´ì „ í•™ìŠµê¸°ì˜ ì•½ì ì„ ë³´ì™„í•˜ë„ë¡ ë§Œë“ ë‹¤.
3. ì”ì—¬ ì˜¤ì°¨ì˜ í™œìš©:
  - ê° í•™ìŠµê¸°ëŠ” ì´ì „ í•™ìŠµê¸°ê°€ ë‚¨ê¸´ ì”ì—¬ ì˜¤ì°¨ì— ëŒ€í•´ í•™ìŠµí•˜ë©°, ì´ ì˜¤ì°¨ë¥¼ ì¤„ì´ëŠ” ë°©í–¥ìœ¼ë¡œ ëª¨ë¸ì„ ì¡°ì •. ì´ë ‡ê²Œ í•˜ì—¬ ëª¨ë¸ì€ ì ì°¨ì ìœ¼ë¡œ ë°ì´í„°ì˜ ë³µì¡í•œ íŒ¨í„´ì„ í•™ìŠµí•  ìˆ˜ ìˆë‹¤.
  - ì”ì—¬ ì˜¤ì°¨ í•™ìŠµì´ë€?
    - GBMì—ì„œ ì‚¬ìš©ë˜ëŠ” ì£¼ìš” ê°œë…ìœ¼ë¡œ, ëª¨ë¸ì´ ë°ì´í„°ì˜ íŒ¨í„´ì„ ì–´ë–»ê²Œ í•™ìŠµí•˜ëŠ”ì§€ì— ëŒ€í•œ ê²ƒì´ë‹¤. ì´ ì ‘ê·¼ë²•ì—ì„œ, ëª¨ë¸ì€ ì´ì „ í•™ìŠµê¸°ì˜ ì˜¤ì°¨(ì”ì—¬ ì˜¤ì°¨)ë¥¼ ëŒ€ìƒìœ¼ë¡œ í•™ìŠµ.
    - ì²« ë²ˆì§¸ ë‹¨ê³„: ì²« ë²ˆì§¸ ì•½í•œ í•™ìŠµê¸°ëŠ” ì›ë³¸ ë°ì´í„°ë¥¼ í•™ìŠµí•˜ì—¬ ì˜ˆì¸¡ì„ ìˆ˜í–‰. ì´ ì˜ˆì¸¡ê³¼ ì‹¤ì œ ê°’ ì‚¬ì´ì˜ ì°¨ì´(ì”ì—¬ ì˜¤ì°¨)ê°€ ê³„ì‚°.
    - ì´í›„ ë‹¨ê³„: ë‹¤ìŒ í•™ìŠµê¸°ëŠ” ì´ì „ í•™ìŠµê¸°ì˜ ì”ì—¬ ì˜¤ì°¨ì— ëŒ€í•´ í•™ìŠµì„ ìˆ˜í–‰. ì¦‰, ì²« ë²ˆì§¸ í•™ìŠµê¸°ê°€ ì˜ëª» ì˜ˆì¸¡í•œ ë¶€ë¶„ì„ ë³´ì™„í•˜ê¸° ìœ„í•´ ë…¸ë ¥.
    - ë°˜ë³µ ê³¼ì •: ì´ ê³¼ì •ì€ ì§€ì •ëœ ìˆ˜ì˜ í•™ìŠµê¸°ê°€ ìƒì„±ë  ë•Œê¹Œì§€ ë˜ëŠ” ë” ì´ìƒì˜ ì„±ëŠ¥ í–¥ìƒì´ ì—†ì„ ë•Œê¹Œì§€ ë°˜ë³µ.
4. ì†ì‹¤ í•¨ìˆ˜ ìµœì í™”:
  - GBMì€ ì£¼ì–´ì§„ ì†ì‹¤ í•¨ìˆ˜(loss function)ë¥¼ ìµœì†Œí™”í•˜ëŠ” ë°©í–¥ìœ¼ë¡œ ëª¨ë¸ì„ ìµœì í™”. ë‹¤ì–‘í•œ ì¢…ë¥˜ì˜ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìœ¼ë©°, ë¬¸ì œì˜ ì¢…ë¥˜(íšŒê·€, ë¶„ë¥˜ ë“±)ì— ë”°ë¼ ì ì ˆí•œ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì„ íƒ.
5. ê³¼ì í•© ë°©ì§€ ê¸°ë²•:
  - GBMì€ í•™ìŠµë¥ (learning rate)ì´ë¼ëŠ” í•˜ì´í¼íŒŒë¼ë¯¸í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ë‹¨ê³„ì˜ í•™ìŠµê¸°ê°€ í•™ìŠµì— ê¸°ì—¬í•˜ëŠ” ì •ë„ë¥¼ ì¡°ì ˆ. ë˜í•œ, íŠ¸ë¦¬ì˜ ê¹Šì´ë‚˜ ë…¸ë“œì˜ ìˆ˜ë¥¼ ì œí•œí•˜ëŠ” ë°©ë²•ìœ¼ë¡œ ê³¼ì í•©ì„ ë°©ì§€í•  ìˆ˜ ìˆë‹¤.

GBMì€ ë†’ì€ ì˜ˆì¸¡ ì„±ëŠ¥ì„ ê°€ì§€ê³  ìˆì§€ë§Œ, í›ˆë ¨ ì‹œê°„ì´ ì˜¤ë˜ ê±¸ë¦¬ê³ , í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹ì´ ì¤‘ìš”í•œ ë‹¨ì ì´ ìˆë‹¤. ìµœê·¼ì—ëŠ” GBMì„ ê¸°ë°˜ìœ¼ë¡œ í•œ ê³ ê¸‰ ì•Œê³ ë¦¬ì¦˜ë“¤ì´ ê°œë°œë˜ì—ˆë‹¤. ëŒ€í‘œì ìœ¼ë¡œ XGBoost, LightGBM, CatBoost ë“±ì´ ìˆìœ¼ë©°, ì´ë“¤ì€ GBMì˜ ê¸°ë³¸ ê°œë…ì„ ìœ ì§€í•˜ë©´ì„œë„ ì„±ëŠ¥, ì†ë„, í™•ì¥ì„± ë©´ì—ì„œ ê°œì„ ëœ ê¸°ëŠ¥ì„ ì œê³µ.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree

# ë°ì´í„° ìƒì„±
X, y = make_classification(n_samples=100, n_features=4, n_informative=2, n_redundant=0, random_state=0)

# ê· í˜• íŠ¸ë¦¬ ë¶„í• 
balanced_tree = DecisionTreeClassifier(max_depth=3, random_state=0)
balanced_tree.fit(X, y)

# ë¦¬í”„ ì¤‘ì‹¬ íŠ¸ë¦¬ ë¶„í• 
# scikit-learnì˜ DecisionTreeClassifierëŠ” ë¦¬í”„ ì¤‘ì‹¬ ë¶„í• ì„ ì§ì ‘ ì§€ì›í•˜ì§€ ì•Šì§€ë§Œ, max_leaf_nodesë¥¼ ì‚¬ìš©í•˜ì—¬ ë¹„ìŠ·í•˜ê²Œ ì‹œë®¬ë ˆì´ì…˜ í•  ìˆ˜ ìˆìŒ
leaf_tree = DecisionTreeClassifier(max_leaf_nodes=8, random_state=0)
leaf_tree.fit(X, y)

# 2ê°œì˜ íŠ¸ë¦¬ ì‹œê°í™”
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
plot_tree(balanced_tree, filled=True, ax=ax[0], feature_names=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
ax[0].set_title('Balanced Tree Split')
plot_tree(leaf_tree, filled=True, ax=ax[1], feature_names=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
ax[1].set_title('Leaf-wise Tree Split')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import seaborn as sns

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
           'hours-per-week', 'native-country', 'income']
data_org = pd.read_csv(url, header=None, names=columns, na_values='?', skipinitialspace=True)
```

```python
data = data_org.copy()

# 2. ê³„ì¸¡ì¹˜ ì²˜ë¦¬
data.dropna(inplace=True)

# 3. ì´ìƒì¹˜ ì œê±° (ì—¬ê¸°ì„œëŠ” 'capital-gain' ê³¼ 'capital-loss'ì—ì„œ ê·¹ë‹¨ì ì¸ ê°’ë“¤ì„ ì´ìƒì¹˜ë¡œ ê°€ì •)
Q1 = data['capital-gain'].quantile(0.25)
Q3 = data['capital-gain'].quantile(0.75)
IQR = Q3 - Q1

capital_gain_outliers = data[(data['capital-gain'] < Q1 - 1.5 * IQR) | (data['capital-gain'] > Q3 + 1.5 * IQR)]
print(capital_gain_outliers)

Q1 = data['capital-loss'].quantile(0.25)
Q3 = data['capital-loss'].quantile(0.75)
IQR = Q3 - Q1

capital_loss_outliers = data[(data['capital-loss'] < Q1 - 1.5 * IQR) | (data['capital-loss'] > Q3 + 1.5 * IQR)]
print(capital_loss_outliers)

data = data.drop(capital_gain_outliers.index)
data = data.drop(capital_loss_outliers.index)

# 4. íŒŒìƒë³€ìˆ˜ ì‘ì„±
data['capital_diff'] = data['capital-gain'] - data['capital-loss']
```

```python
# 'age' ì»¬ëŸ¼ì˜ ì´ìƒì¹˜ë¥¼ í‰ê· ê°’ìœ¼ë¡œ ëŒ€ì²´
age_mean = data['age'].mean()
data['age'] = data['age'].apply(lambda x : x if x <= 74 else age_mean)

# 'education-num' ì»¬ëŸ¼ì˜ ì´ìƒì¹˜ë¥¼ í‰ê· ê°’ìœ¼ë¡œ ëŒ€ì²´
education_num_mean = data['education-num'].mean()
data['education-num'] = data['education-num'].apply(lambda x : x if x >= 5 else education_num_mean)
```

```python
# ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”©
data['sex'] = data['sex'].map({'Male': 0, 'Female': 1})

data['income'] = data['income'].map({'<=50K': 0, '>50K': 1})
```

```python
# ì•ˆì“°ëŠ” ì»¬ëŸ¼ ë“œë
data = data.drop(['fnlwgt'], axis=1)

data = data.drop(['education'], axis=1)
data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


# ëª¨ë¸ í‰ê°€
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # ì˜ˆì¸¡
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")
    print(f"Confusion_matrix: {conf_matrix}")
    print(f"Classification_report: {class_report}")

    # ì‹œê°í™”
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Confusion Matrix - {model}")
    plt.xlabel(model)
    plt.tight_layout()
    plt.show()
    print()

X = data.drop('income', axis=1)
y = data['income']

# 7. í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 8. ë°ì´í„° í‘œì¤€í™”
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

hyperparameters = {
    'n_estimators':[100, 150, 200, 250, 300],
    'learning_rate': [0.1, 0.3, 0.5, 0.7, 1],
    'max_depth': [2, 3, 4, 5],
    'random_state':[0, 10, 42, 100, 156]
}

# ëª¨ë¸ ìƒì„±
xgb_wrapper = XGBClassifier()


# GridSearchCV ì„¤ì • ë° ì ìš©
grid_search = GridSearchCV(estimator=xgb_wrapper,
                           param_grid=hyperparameters,
                           n_jobs=-1,
                           verbose=2,
                           cv=5)

grid_search.fit(X_train, y_train)

# ìµœì ì˜ í•˜ì´í¼íŒŒë¼ë¯¸í„° ì¶œë ¥
print(f"Best parameters: {grid_search.best_params_}")

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_model = grid_search.best_estimator_
evaluate_model(best_model, X_train, y_train, X_test, y_test)
```

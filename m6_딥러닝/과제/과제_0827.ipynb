Task1_0827. 2행 3열 텐서의 크기를 3행 2열로 변경하세요.
- view 사용
- reshape 사용
- transpose 사용

```python
import torch

x = torch.rand(2, 3)
print(x)
```

```python
# view
y = x.view(3, 2)
print(y)
```

```python
# reshape
y = x.reshape(3, 2)
print(y)
```

```python
# transpose
y = x.T
print(y)

z = x.transpose(0, 1)
print(z)
```

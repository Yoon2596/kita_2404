```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup

# 데이터를 저장할 리스트
names = []
sections = []
prices = []
addresses = []
rates = []
intros = []
# reviews = {}

query = 1
url = 'https://www.menupan.com/Restaurant/search/search_main.asp?page={}&trec=206&areacode=ss218'

driver = webdriver.Chrome()

try:
    driver.get(url.format(query))

    time.sleep(1)

    # restName 클래스의 a 태그를 선택
    rest_elements = driver.find_elements(By.CSS_SELECTOR, '.restName a')

    for rest_element in rest_elements:
        # 클릭하여 새 탭에서 링크를 엽니다
        rest_element.click()
        time.sleep(2)  

        # 새 탭으로 전환
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(2)

        # 상세 페이지에서 필요한 정보 추출
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # 이름 추출
        name_item = soup.select_one('dd.name')
        if name_item:
            name_text = name_item.get_text(strip=True)
            name_text = name_text.split('[', 1)[0].strip()
            names.append(name_text)
            print(f'Name: {name_text}')
        
        # 섹션
        section_item = soup.select_one('dd.type')
        if section_item:
            section_text = section_item.get_text(strip=True)
            sections.append(section_text)
            print(f'Section: {section_text}')

        # 가격
        price_item = soup.select_one('.price')
        if price_item:
            price_item = price_item.get_text(strip=True)
            prices.append(price_item)
            print(f'Section: {price_item}')

        # 주소
        address_item1 = soup.select_one('dd.add1') 
        address_item2 = soup.select_one('dd.add2') 
        if address_item2:
            address_text = address_item2.get_text(strip=True)
            address_text = address_text.split(' ', 1)[1].strip()
            addresses.append(address_text)
            print(f'Address: {address_text}')
        elif address_item1 == True and address_item2 != True: 
            address_text = address_item1.get_text(strip=True)
            addresses.append(address_text)
            print(f'Address: {address_text}')

        # 평가 점수
        rate_item = soup.select_one('span.total')
        if rate_item:
            rate_text = rate_item.get_text(strip=True)
            rates.append(rate_text)
            print(f'Rate: {rate_text}')

        # 내용
        intro_item = soup.select_one('.article')
        if intro_item:
            intro_text = intro_item.get_text(strip=True)
            intros.append(intro_text)
            print(f'Intro: {intro_text}')

        break
        # review_item = soup.select_one('dd.content')
        # if review_item:
        #     intro_text = intro_item.get_text(strip=True)
        #     intros.append(intro_text)
        #     print(f'Intro: {intro_text}')

        # 새 탭을 닫고 원래 탭으로 돌아가기
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(3)  # 이전 페이지가 로드될 시간을 줍니다

finally:
    driver.quit()
```

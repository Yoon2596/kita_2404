```python
file_path = 'catchtable_3.csv'
dl = pd.read_csv(file_path)
grouped_dl = dl.groupby('shop_name').first().reset_index()
```

```python
file_path = 'sungsoo_keyword_0704.csv'
df = pd.read_csv(file_path)
df.rename(columns={'Name': 'shop_name'}, inplace=True)
```

```python
# 머지 1
merged_df = pd.merge(df, grouped_dl, on='shop_name', how='inner')
print(merged_df)
merged_df.rename(columns={'shop_name': 'Name'}, inplace=True)
```

```python
#머지 2
df = pd.merge(merged_df, data, on='Name', how='inner')

df.to_excel('0704_sungdong.xlsx')
df.to_csv('0704_sungdong.csv')
```

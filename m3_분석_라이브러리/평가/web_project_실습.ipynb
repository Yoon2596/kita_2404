```python
import pandas as pd

# 데이터 파일 로드
# file_path = 'catchtable_3.csv'
file_path = 'practice2.csv'
data = pd.read_csv(file_path)

df = data.copy()

# # avg_lunch와 avg_dinner 컬럼에서 NA 값을 0으로 대체
df['avg_lunch'] = df['avg_lunch'].fillna(0)
df['avg_dinner'] = df['avg_dinner'].fillna(0)

# avg_lunch와 avg_dinner 컬럼에서 '영업안함'과 '가격변동' 값을 0으로 대체
df['avg_lunch'] = df['avg_lunch'].replace(['영업안함', '가격변동'], 0)
df['avg_dinner'] = df['avg_dinner'].replace(['영업안함', '가격변동'], 0)

# avg_lunch와 avg_dinner 컬럼을 float 타입으로 변환
df['avg_lunch'] = df['avg_lunch'].astype(float)
df['avg_dinner'] = df['avg_dinner'].astype(float)

# 데이터 프레임 열 이름 확인
# print("데이터 프레임의 열 이름:")
# print(df.columns)

category_1_region = {
    1: '강남구',
    2: '마포구',
    3: '용산구',
    4: '성동구'
}

category_2_select = {
    1: '음식 카테고리',
    2: '가격대',
    3: '키워드'
}

category_3_mapping = {
    1: '한식',
    2: '양식',
    3: '일식',
    4: '중식',
    5: '아시아음식',
    6: '퓨전',
    7: '뷔페',
    8: '주류',
    9: '카페'
}

top_keywords = ['특별한 날', '분위기 좋은', '친절한', '음식이 맛있는', '서비스 좋은', '데이트/연인과 함께', '주류', '추천하는']

def search_by_category():
    # Category_1 목록과 번호를 출력
    print("---------------")
    print("검색할 구역 목록입니다:")
    for number, category in category_1_region.items():
        print(f"{number}: {category}")
    print("---------------")

    category_region_num = int(input("검색할 구역을 입력하세요: (1: 강남구, 2: 마포구, 3: 용산구, 4: 성동구) : "))
    selected_region = category_1_region[category_region_num]
    print(f"선택한 구역 : {selected_region}")

    # 선택한 구역에 해당하는 데이터 필터링
    selected_data = df[df['Location_1'] == selected_region]

    # Category_2 목록과 번호를 출력
    print("---------------")
    print("검색 주제입니다:")
    for number, category in category_2_select.items():
        print(f"{number}: {category}")
    print("---------------")

    selected_num = int(input("검색할 주제를 입력하세요: "))
    selected = category_2_select[selected_num]
    print(f"선택한 주제 : {selected}")

    if selected_num == 1:
        # Case 1: 음식 카테고리
        print("---------------")
        print("검색할 음식 카테고리 목록:")
        for number, category in category_3_mapping.items():
            print(f"{number}: {category}", end=' ')
        print("---------------")

        # 사용자로부터 음식 카테고리 선택 입력 받기
        category_3_number = int(input("검색할 음식 카테고리를 입력하세요: "))
        selected_category_1 = category_3_mapping[category_3_number]
        print(f"선택한 Category_1: {selected_category_1}", end=' ')

        # 선택한 Category_1에 해당하는 세부 Category 목록 출력
        sub_category_list = selected_data[selected_data['Category_1'] == selected_category_1]['Category'].unique()
        print("---------------")
        print("세부 Category 목록:")
        for idx, category in enumerate(sub_category_list):
            print(f"{idx + 1}. {category}")
        print("---------------")

        # 사용자로부터 세부 Category 선택 입력 받기
        sub_category_idx = int(input("검색할 세부 음식 주제를 입력하세요: ")) - 1
        selected_sub_category = sub_category_list[sub_category_idx]
        print(f"선택한 세부 Category: {selected_sub_category}")

        # 선택한 Category_1과 세부 Category에 해당하는 데이터 필터링
        result = selected_data[(selected_data['Category_1'] == selected_category_1) & (selected_data['Category'] == selected_sub_category)]
    
    elif selected_num == 2:
        # Case 2: 가격대
        while True:
            try:
                meal = input('점심 또는 저녁을 선택해주세요 (점심/저녁): ').strip()
                if meal not in ['점심', '저녁']:
                    print("올바른 선택을 해주세요 (점심 또는 저녁).")
                    continue
                
                lower_budget = int(input('현재 원하시는 최소 가격대를 적어주세요 (최소 1만원): ').strip())
                upper_budget = int(input('현재 원하시는 최대 가격대를 적어주세요 (최대 40만원): ').strip())
                
                # 조건 확인
                if lower_budget >= 1 and upper_budget <= 40 and upper_budget >= lower_budget:
                    lower_budget *= 10000
                    upper_budget *= 10000
                    print(f"최소 가격: {lower_budget}원")
                    print(f"최대 가격: {upper_budget}원")
                    break
                else:
                    print("가격대가 올바르지 않습니다. 다시 입력해주세요.")
            except ValueError:
                print("숫자를 입력해주세요.")

        # 범위에 따라 데이터를 필터링
        if meal == '점심':
            result = selected_data[(selected_data['avg_lunch'] >= lower_budget) & (selected_data['avg_lunch'] <= upper_budget) | 
                        (selected_data['min_lunch'] >= lower_budget) & (selected_data['max_lunch'] <= upper_budget)]
        elif meal == '저녁':
            result = selected_data[(selected_data['avg_dinner'] >= lower_budget) & (selected_data['avg_dinner'] <= upper_budget) | 
                        (selected_data['min_dinner'] >= lower_budget) & (selected_data['max_dinner'] <= upper_budget)]
        else:
            result = pd.DataFrame()
            print('해당하는 결과의 음식점은 없습니다')

    elif selected_num == 3:
        # Case 3 : 키워드
        print("---------------")
        print("키워드 목록:")
        for idx, keyword in enumerate(top_keywords):
            print(f"{idx + 1}. {keyword}")
        print("---------------")

        # 사용자로부터 키워드 선택 입력 받기
        keyword_idx = int(input("검색할 키워드를 선택하세요 (번호로 입력): ")) - 1
        selected_keyword = top_keywords[keyword_idx]
        print(f"선택한 키워드: {selected_keyword}")

        # 선택한 키워드에 해당하는 데이터 필터링
        result = selected_data[selected_data['Top Keywords List'].str.contains(selected_keyword, na=False)]
    else:
        result = pd.DataFrame()
        print("잘못된 선택입니다.")

    while True:
        # 사용자로부터 추천받을 식당 수 입력 받기
        num_recommendations = int(input("식당 몇 개를 추천할까요? "))
        
        if len(result) < num_recommendations:
            recommendations = result
        else:
            recommendations = result.nlargest(num_recommendations, 'Rating')

        # 추천 식당 리스트 출력
        print("추천 식당 리스트:")
        print(recommendations[['Unnamed: 0.1', 'Unnamed: 0', 'Name', 'Keywords (10)',
       'Top Keywords List', 'Top Keywords', 'region', 'Price_range', 'lun_di',
       'save', 'Category', 'Category_1', 'Location_y', 'Location_1', 'Label',
       'Rating', 'Review', 'avg_lunch', 'avg_dinner', 'min_lunch', 'max_lunch',
       'min_dinner', 'max_dinner', 'region_1']])

        # 사용자로부터 마음에 드는 식당이 있는지 확인
        satisfied = input("마음에 드시는 식당이 있습니까? (y/n): ")
        if satisfied.lower() == 'y':
            print("좋은 식당을 찾으셨길 바랍니다!")
            break
        else:
            print("다시 추천해드리겠습니다.")
    
    return recommendations

# 검색 함수 실행 및 데이터 프레임으로 결과 저장
search_results = search_by_category()

# 검색 결과 데이터 프레임 표시
print(search_results)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from math import sin, cos, sqrt, atan2, radians

# 파이썬 geopy 라이브러리를 이용해
# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo_local = Nominatim(user_agent='South Korea')  # 지역설정
        location = geo_local.geocode(address)
        if location:
            geo = [location.latitude, location.longitude]
        else:
            geo = [0, 0]
        return geo
    except:
        return [0, 0]

# 두 위도와 경도의 거리를 구하는 공식 (허버사인 공식)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # 지구의 반지름 (킬로미터 단위)

    # 각도를 라디안으로 변환
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance

# 사용자로부터 입력 받기
my_location = input('현재 내가 있는 곳을 알려주세요: ')

# 입력된 주소를 사용하여 위도와 경도 찾기
latitude, longitude = geocoding(my_location)

# 위도와 경도 확인
print(f"내 위치: 위도 {latitude}, 경도 {longitude}")

# search_results 데이터 프레임의 주소 컬럼
address_column = 'region_1' 

# search_results 데이터 프레임에 latitude와 longitude 컬럼 추가
search_results['latitude'] = search_results[address_column].apply(lambda x: geocoding(x)[0])
search_results['longitude'] = search_results[address_column].apply(lambda x: geocoding(x)[1])

# 거리 계산
distances = []
for i in range(len(search_results)):
    lat2 = search_results['latitude'].iloc[i]
    lon2 = search_results['longitude'].iloc[i]
    distance = haversine_distance(latitude, longitude, lat2, lon2)
    distances.append(round(distance, 2))
    print(f"거리 ({latitude}, {longitude}) -> ({lat2}, {lon2}): {distance:.2f} km")

# 계산된 거리 출력
print(distances)

# 검색 결과 데이터 프레임에 거리 정보 추가
search_results['distance'] = distances

# 결과 출력
print(search_results[['Name', 'region_1', 'latitude', 'longitude', 'distance']])
```

```python
# 내가 있는곳과 음식점 주소를 통해 위도와 경도를 계산하여 둘 사이의 거리 계산

import pandas as pd
from geopy.geocoders import Nominatim
from math import sin, cos, sqrt, atan2, radians

# 파이썬 geopy 라이브러리를 이용해
# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo_local = Nominatim(user_agent='South Korea')  # 지역설정
        location = geo_local.geocode(address)
        geo = [location.latitude, location.longitude]
        return geo
    except:
        return [0, 0]

# 사용자로부터 입력 받기
my_location = input('현재 내가 있는 곳을 알려주세요: ')

# 빈 DataFrame 생성
my_loc = pd.DataFrame(columns=['latitude', 'longitude'])

# 입력된 주소를 사용하여 위도와 경도 찾기
latitude, longitude = geocoding(my_location)

# DataFrame에 값 추가
my_loc.loc[0] = [latitude, longitude]

print(my_loc)

# 두 위도와 경도의 거리를 구하는 공식 (허버사인 공식)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # 지구의 반지름 (킬로미터 단위)

    # 각도를 라디안으로 변환
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c

    return distance

# my_loc DataFrame에서 위도와 경도 추출
lat1 = my_loc['latitude'].values[0]
lon1 = my_loc['longitude'].values[0]
print(lat1, lon1)


for idx, addr in enumerate(search_results['Location_2']):
    search_results.loc[idx, 'latitude'] = geocoding(addr)[0]
    search_results.loc[idx, 'longitude'] = geocoding(addr)[1]

# 거리 계산
distances = []
for i in range(len(search_results)):
    lat2 = search_results['latitude'].iloc[i]
    lon2 = search_results['longitude'].iloc[i]
    distance = haversine_distance(lat1, lon1, lat2, lon2)
    distances.append(round(distance, 2))
    print(f"거리 ({lat1}, {lon1}) -> ({lat2}, {lon2}): {distance:.2f} km")

# 계산된 거리 출력
print(distances)
```

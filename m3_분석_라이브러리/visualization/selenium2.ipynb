```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
from lxml import html

url = 'https://web.joongna.com/search/%EC%95%84%EC%9D%B4%ED%8F%B0x?page=11'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)



# 페이지에 있는 소스 가져오기
page_source = driver.page_source
tree = html.fromstring(page_source)

data = []
articles = tree.xpath('//*[@id="__next"]/div/main/div/div/ul/li')
for article in articles:
    try:
        title = article.xpath('.//h2/text()')[0].strip()
        price = article.xpath('.//div[contains(@class, "font-semibold")]/text()')[0].strip()
        spans = article.xpath('.//div[@class="my-1"]/span/text()')

        # span 리스트 출력
        print('span : ', spans)

        # 날짜 텍스트만 출력
        time_diff = spans[-1].strip() if len(spans) > 2 else ""
        data.append({"Title": title, "Price": price, "Time Difference": time_diff})

    except IndexError:
        continue
driver.quit()

data
    
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
from lxml import html

url = 'https://web.joongna.com/search/%EC%95%84%EC%9D%B4%ED%8F%B0x?page=11'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)



# 페이지에 있는 소스 가져오기
page_source = driver.page_source
# tree = html.fromstring(page_source)
soup = BeautifulSoup(page_source, 'html.parser')

data = []
#__next > div > main > div > div > ul > li
articles = soup.select('#__next > div > main > div > div > ul > li')
for article in articles:
    try:
        title = soup.select_one('h2').get_text(strip=True)
        price = soup.select_one('div.font-semibold.text-heading').get_text(strip=True)
        spans = soup.select('div.my-1 > span.text-sm.text-gray-400')

        # 디버깅 : 모든 span 요소 출력
        # for idx, span in enumerate(spans):
            # print('{idx}, {span.get_text(strip=True)}')

        # 날짜 텍스트만 출력
        time_diff = spans[-1].get_text(strip=True) if spans else ""
        data.append({"Title": title, "Price": price, "Time Difference": time_diff})

    except IndexError:
        continue
driver.quit()

data

url = 'https://web.joongna.com/search/%EC%95%84%EC%9D%B4%ED%8F%B0x?page=1'
driver = webdriver.Chrome()
driver.get(url)

# 페이지 소스 가져오기
page_source = driver.page_source
tree = html.fromstring(page_source)

# li 요소들 추출
articles = soup.select('#__next > div > main > div > div > ul > li')
for article in articles:
    try:
        title = article.select_one('h2').get_text()
        price = article.select_one('div.font-semibold.text-heading').get_text(strip=True)
        spans = article.select('div.my-1 > span.text-sm text-gray-400')

        time_diff = spans[-1].get_text(strip=True) if spans else ""
        
        data.append({"Title": title, "Price": price, "Time Difference": time_diff})

    except IndexError:
        continue
driver.quit()
data
```

```python
# csv 파일 읽고 저장하기
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))
# df.to_csv('test.csv')
df.to_csv('test.csv', index=None)       # 이렇게 쓰면 됨 (위에거 쓰면 불러올때 index_col = 0 계속 붙여야함)
```

```python
df1 = df.copy()
df1.to_csv('test1.csv')
df1 = pd.read_csv('test1.csv', index_col=0)
df1
```

```python
# 인덱싱으로 내용 바꿈
df1.iloc[0] = 2
df1.iloc[1] = 3
df1
```

```python
# json 파일 읽고 쓰기
df1.to_json('test2.json')
df_j = pd.read_json('test2.json')
df_j
```

```python
# 바이너리 파일 읽고 쓰기
df1.to_pickle('test3.pkl')
df1 = pd.read_pickle('test3.pkl')
df1
```

```python
# 간단한 데이터프레임 작성 후 csv, json, pickle 형태로 다시 불러와서 출력하는 연습
# cvs 파일 생성 및 읽기
import numpy as np
import pandas as pd

# 2x5 매트릭스 만들기
ar2 = np.arange(10).reshape(2, 5)
# 데이터프레임 생성
df = pd.DataFrame(ar2, columns=list('abcde'))
# 데이터프레임을 csv 파일로 만듬
df.to_csv('prac.csv', index = None)
```

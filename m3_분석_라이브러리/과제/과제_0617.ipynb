Task1_0617. 선그래프, 히스토그램, 산점도, 막대그래프 4개의 subplot으로 구성된 2 x 2 그래프를 작성하세요.

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# x = np.arange(1, 21, 2)
# y = np.random.randint(1, 10, 10)

ax1.plot(np.random.randn(20).cumsum(), 'k--', marker='o', markersize = 3)
ax2.hist(np.random.rand(1000), bins=20, color='k', alpha=0.5)
ax3.scatter(np.arange(30), np.arange(30)+3 * np.random.randn(30))
ax4.bar(np.random.randn(50), np.random.rand(50))

plt.show()
```

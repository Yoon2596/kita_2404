```python
#Q. 다음 사항을 수행하세요
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 평균 50, 표준편차 10인 정규 분포에서 100개 샘플 생성
np.random.seed(0)
datas = np.random.normal(50, 10, 100)       #(평균, 표준편차10, 갯수)
# print(datas)

# 기본 데이터의 평균가 중앙값 계산
# data = np.array(datas)
mean = np.mean(datas)
median = np.median(datas)
print(f'Mean : {mean}, Median : {median}')

# 데이터에 이상치 500 추가
new_data = np.append(datas, 500)
# print(new_data)
# 이상치가 포함 된 데이터의 평균과 중앙값 계산\
new_mean = np.mean(new_data)
new_median = np.median(new_data)
print(f'New_Mean : {new_mean}, New_Median : {new_median}')


# 평균과 중간값 차이
print(new_mean - mean, new_median - median)

# 이상치 포함에 따른 평균값과 중앙값간의 변화 설명
# 이상치 500이 포함된 중앙값이 기존 중앙값보다 0.3정도 증가 한 것을 확인
# print(data)
```

```python
import numpy as np
import pandas as pd

# 시드 설정 및 데이터 생성
np.random.seed(0)
datas = np.random.normal(50, 10, 100)  # (평균 50, 표준편차 10, 갯수 100)

# 데이터프레임 생성
data = pd.DataFrame(datas, columns=['sample'])

# mean과 median 값을 새로운 행으로 추가
mean_value = data['sample'].mean()
median_value = data['sample'].median()

# 새로운 행 생성
mean_row = pd.DataFrame({'sample': [mean_value], 'statistic': ['mean']})
median_row = pd.DataFrame({'sample': [median_value], 'statistic': ['median']})

# 기존 데이터프레임에 새로운 행 추가
data['statistic'] = 'sample'  # 기존 데이터프레임에 'statistic' 컬럼 추가
data = pd.concat([data, mean_row, median_row], ignore_index=True)

print(data.tail(12))  # 마지막 12개의 행 출력
```

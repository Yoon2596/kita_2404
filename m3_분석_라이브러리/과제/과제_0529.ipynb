Task7_0529. 주어진 DataFrame의 특정 열에 대해 map을 사용하여 등급을 부여하는 함수를 적용하여 변환을 수행하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Score": [85, 90, 95, 80, 75]
})
```

```python
# 성적에 따라 등급을 부여하는 함수 정의
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    else:
        return 'C'


# score 열에 대해 등급을 부여하는 함수
df['Grade'] = df['Score'].map(grade)

#결과 출력
print(df)
```

Task12_0529. 주어진 DataFrame의 특정 열에 대해 제곱과 제곱근을 계산하는 함수들을 동시에 적용하여 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "Numbers": [1, 2, 3, 4, 5]
})
df
```

```python
# 여러 함수를 적용하여 새로운 열 생성 (제곱과 제곱근)
df['Squared'] = df['Numbers'].apply(lambda x: x ** 2)
df['Squared_Root'] = df['Numbers'].apply(lambda x: x ** 0.5)

print(df)
```

```python
# 다른 방식으로 푼 경우 (agg)

import pandas as pd
import numpy as np

# 데이터프레임 생성
df = pd.DataFrame({
    "Numbers": [1, 2, 3, 4, 5]
})

def pow(x):
  return x**2

def sqrt(x):
  return np.sqrt(x)

res = df.Numbers.agg([pow, sqrt])
df = pd.concat([df,res], axis=1)
df
```

```python
import pandas as pd
import numpy as np

# 데이터프레임 생성
df = pd.DataFrame({
    "Numbers": [1, 2, 3, 4, 5]
})

result = df['Numbers'].agg([lambda x: x**2, lambda x: x ** 0.5])
result.columns = ['Squared', 'Squared_Root']
df = df.join(result)
df
```

```python
# 다른 방식으로 푼 경우 (튜플을 시리즈로 변환해서 푼 경우)

def SQ_SQRT(x):
  return round(x*x,2), round(x**0.5,2)

# Option1
df[['SQ', 'SQRT']] =df['Numbers'].transform(lambda x: pd.Series(SQ_SQRT(x)))        ## Apply나 Transform이나 동일
print(df,"\n")

# Option2
df[['SQ', 'SQRT']] = df['Numbers'].apply(lambda x: (round(x*x,2), round(x**0.5,2))).apply(pd.Series)
print(df,"\n")
```

Task14_0529. 주어진 데이터프레임에서 각 그룹별로 Value 열을 표준화하여 새로운 열로 추가하여 출력하세요.(apply, transform 두가지 적용)

```python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B', 'B', 'A', 'B', 'A'],
    'Value': [10, 15, 10, 20, 10, 30, 50, 10, 20, 15]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# 그룹별 표준화하는 함수 정의
def standardize(x):
    return (x - x.mean()) / x.std()
```

```python
# apply를 사용하여 그룹별 표준화 (추가 처리 필요)
standardized_series = df.groupby('Category')['Value'].apply(standardize)
print(standardized_series)
```

```python
# 그룹별 표준화하는 함수 정의
def standardize(x):
    return (x - x.mean()) / x.std()

# 각 그룹별로 Value 열을 표준화하여 새로운 열로 추가
df['Standardized_Value'] = df.groupby('Category')['Value'].transform(standardize)
print(df)
```

```python
## Option-1 Using TRANSFORM
df["그룹평균"]=df.groupby("Category")["Value"].transform(lambda x: round(x.mean(),2))
df["STD"]=df.groupby("Category")["Value"].transform(lambda x: round(x.std(),2))
df["Z-Score"]=df.groupby("Category")["Value"].transform(standardize)
#print(df,'\n')

## Option-2 Using MAP

CatMean=round(df.groupby("Category")["Value"].mean(),2)
CatStd=round(df.groupby("Category")["Value"].std(),2)

df["그룹평균2"]=df["Category"].map(CatMean)
df["STD2"]=df["Category"].map(CatStd)
df["Z-Score2"] = round((df["Value"] - df["그룹평균2"]) / df["STD2"],2)
#print(df,'\n')

## Option-3 Using APPLY
def M_S_Z(df_g):
    df_g['그룹평균3'] = round(df_g["Value"].mean(),2)
    df_g['STD3'] = round(df_g["Value"].std(),2)
    df_g['Z-Score3'] = round((df_g["Value"] - df_g["그룹평균3"]) / df_g['STD3'],2)
    return df_g
df = df.groupby('Category', group_keys=False).apply(M_S_Z)
df
```

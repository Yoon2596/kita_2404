```python
a = pd.read_excel('용산구.xlsx')
b = pd.read_excel('용산구_가게만.xlsx')

# Let's assume you need to insert b into the first 8 columns of a.
required_columns = 8

# If b has fewer columns, add empty columns to match the required number.
if b.shape[1] < required_columns:
    for _ in range(required_columns - b.shape[1]):
        b[f'Extra_Column_{b.shape[1] + _}'] = ""


# Define the interval at which you want to insert rows from `b` into `a`
interval = 50

# Iterate over DataFrame `b` and insert its rows into `a`
for i, row in b.iterrows():
    # Calculate the position where the row from `b` should be inserted into `a`
    insert_position = i * interval
    
    # Insert the row from `b` into the first `required_columns` columns in `a`
    for j in range(interval):
        if insert_position + j < len(a):
            a.iloc[insert_position + j, :required_columns] = row.values

# Save the modified DataFrame back to an Excel file if needed
a.to_excel('modified_용산구.xlsx', index=False)
```
